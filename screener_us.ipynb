# Download Data

```python
def get_data_batch(tickers):
    df = yf.download(
        tickers=tickers,
        period="6mo",
        interval="1d",
        group_by="ticker",
        threads=True,
        progress=False
    )
    return df

# Download
data = get_data_batch(tickers)
print(f"Total downloaded data: {len(data)}")
```

# Logic Calculation

```python
def calculate(df):
    df = df.copy()

    # ===== MOVING AVERAGES =====
    df["MA3"] = df["Close"].rolling(3).mean()
    df["MA5"] = df["Close"].rolling(5).mean()
    df["MA10"] = df["Close"].rolling(10).mean()
    df["MA20"] = df["Close"].rolling(20).mean()

    # ===== PREV CLOSE =====
    df["PrevClose"] = df["Close"].shift(1)

    # ===== TREND =====
    df["AboveMA"] = (
        (df["Close"] > df["MA3"]) &
        (df["Close"] > df["MA5"]) &
        (df["Close"] > df["MA10"]) &
        (df["Close"] > df["MA20"])
    )

    df["DistanceMA20"] = (df["Close"] - df["MA20"]) / df["MA20"] * 100

    df["BUY_SETUP"] = (
        df["AboveMA"] &
        (df["DistanceMA20"] <= 5)
    )

    # ===== VOLUME =====
    df["VolumeMA20"] = df["Volume"].rolling(20).mean()
    df["VolumeRatio"] = df["Volume"] / df["VolumeMA20"]

    # ===== ATR =====
    high_low = df["High"] - df["Low"]
    high_close = (df["High"] - df["Close"].shift(1)).abs()
    low_close = (df["Low"] - df["Close"].shift(1)).abs()

    tr = pd.concat([high_low, high_close, low_close], axis=1).max(axis=1)
    df["ATR14"] = tr.rolling(14).mean()

    return df
```

# Helper Format

```python
def fmt_volume(volume, ratio, open_price, close_price):

    if volume >= 1e9:
        v = f"{volume/1e9:.2f}B"
    elif volume >= 1e6:
        v = f"{volume/1e6:.2f}M"
    else:
        v = f"{int(volume):,}"

    # ===== candle color proxy =====
    if close_price > open_price:
        emoji = "🟢"
        direction = "BUY!"
    elif close_price < open_price:
        emoji = "🔴"
        direction = "SELL!"
    else:
        emoji = "⚪"
        direction = "Neutral"

    return f"{emoji} {v} ({ratio:.2f}x) [{direction}]"

def fmt_ma(price, ma):
    if np.isnan(ma):
        return "N/A"

    diff_pct = (ma - price) / price * 100

    if diff_pct <= 0:
        emoji = "🟢"
    else:
        emoji = "🔴"

    return f"{ma:.2f} ({emoji}{diff_pct:+.2f}%)"
```

# Scanning Data

```python
results = []

for ticker in tickers:
    try:
        if ticker not in data.columns.get_level_values(0):
            continue

        df = data[ticker].dropna()

        if len(df) < 30:
            continue

        df = calculate(df)
        today = df.iloc[-1]

        if not bool(today["BUY_SETUP"]):
            continue

        # ===== PRICE DATA =====
        price = float(today["Close"])
        high = float(today["High"])
        low = float(today["Low"])
        prev_close = float(today["PrevClose"])

        # ===== CHANGE PCT =====
        change_pct = ((price - prev_close) / prev_close) * 100

        # ===== ATR =====
        atr = float(today["ATR14"])

        # ===== RESISTANCE =====
        next_resistance = high if high > price else np.nan

        # ===== TP LOGIC (HYBRID) =====
        if not np.isnan(next_resistance):

            tp_mode = "RESISTANCE"
            tp1 = next_resistance
            tp2 = next_resistance * 1.02

            reward1 = (tp1 - price) / price * 100
            reward2 = (tp2 - price) / price * 100

        else:

            tp_mode = "ATR"
            tp1 = price + atr
            tp2 = price + (2 * atr)

            reward1 = atr / price * 100
            reward2 = (2 * atr) / price * 100

        # ===== STORE RESULT =====
        results.append({
            "Ticker": ticker,

            "Price": price,
            "High": high,
            "Low": low,
            "PrevClose": prev_close,
            "ChangePct": change_pct,
            "Open": float(today["Open"]),

            "Volume": float(today["Volume"]),
            "VolumeRatio": float(today["VolumeRatio"]),

            "MA3": float(today["MA3"]),
            "MA5": float(today["MA5"]),
            "MA10": float(today["MA10"]),
            "MA20": float(today["MA20"]),

            "TPMode": tp_mode,
            "TP1": tp1,
            "TP2": tp2,

            "RewardPct1": reward1,
            "RewardPct2": reward2,
        })

    except Exception:
        continue
```

# Output Table

```python
result = pd.DataFrame(results)

if result.empty:
    print("No signals found")
    raise SystemExit

result = result.sort_values(by="VolumeRatio", ascending=False)

print(f"SUPERKETAT FOUND: {len(result)}")

print(
    tabulate(
        result,
        headers="keys",
        tablefmt="pretty",
        showindex=False
    )
)
```

# Telegram Formatting

```python
def format_signal_row(row):

    price = row["Price"]
    high = row["High"]
    low = row["Low"]
    change_pct = row["ChangePct"]

    # ===== emoji =====
    if change_pct >= 0:
        emoji = "🟢"
        sign = "+"
    else:
        emoji = "🔴"
        sign = ""

    # ===== TP TEXT =====
    if row["TPMode"] == "ATR":
        tp_text = (
            f"TP1 (1 ATR): {row['TP1']:.2f} (+{row['RewardPct1']:.2f}%)\n"
            f"TP2 (2 ATR): {row['TP2']:.2f} (+{row['RewardPct2']:.2f}%)"
        )
    else:
        tp_text = (
            f"TP1 : {row['TP1']:.2f} (+{row['RewardPct1']:.2f}%)\n"
            f"TP2 : {row['TP2']:.2f} (+{row['RewardPct2']:.2f}%)"
        )

    return f"""
STOCKS: {row['Ticker']}
══════════════════════

💰 Price Action
{emoji} Close: {price:.2f} ({sign}{change_pct:.2f}%)
📈 High: {high:.2f} | Low: {low:.2f}
📊 Volume: {fmt_volume(row['Volume'], row['VolumeRatio'], row['Open'], row['Price'])}

📉 Moving Average Status
MA3  : {fmt_ma(price, row['MA3'])}
MA5  : {fmt_ma(price, row['MA5'])}
MA10 : {fmt_ma(price, row['MA10'])}
MA20 : {fmt_ma(price, row['MA20'])}

🎯 Take Profit
{tp_text}

━━━━━━━━━━━━━━━━━━
"""
```
